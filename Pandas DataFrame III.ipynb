# .apply() method overview

```python
bond = pd.read_csv("Dataset/jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def addMillion(number):
    return str(number) + " million"
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for column in columns:
    bond[column] = bond[column].apply(addMillion)
    
bond.head(2)
```

```python
bond.head()
```

```python
def movies_review(movie):
    box_office = movie[3]
    budget = movie[4]
    profit = float(box_office - budget)
    
    if profit > 500:
        return "Smashing hit!!!"
    elif profit > 300 and profit < 500:
        return "Great hit!!!"
    elif profit > 100 and profit <300:
        return "Not bad!!!"
    else:
        return "What the frack!!!"
```

```python
bond.apply(movies_review,axis="columns")
```
